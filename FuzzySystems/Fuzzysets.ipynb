# **Fuzzy Sets**

**I. Introduction**

Fuzzy sets are an extension of crisp (two-valued) sets to handle the concept of partial truth, which enables the modeling of the uncertainties of natural langauges.

Eg: when it is **very** cloudy, it will **most probably** rain

**II. Formal Definitions**

Suppose $X$ is the domain, or universe of discourse, and $x \in X$ is a specific element of the domain $X$. Then, the fuzzy set $A$ is characterized by a membership mapping function:

$ μ_A:X \to [0,1] $

Thefore, for all $x \in X, \mu_A(x)$ indicates the certainty to which element $x$ belongs to fuzzy set $A$. For two-valued sets, $μ_A(x)$ is either 0 or 1.

Fuzzy sets can be defined for **discrete** (finite) or **continuous** (infinite) domains. The notation used to denote fuzzy sets differ based on the type of domain over which that set is defined.

In the case of a discrete domain $X$, the fuzzy set can either be expressed in the form of an $n_x-dimensional$ vector or using the sum notation.

If $X=\{ x_1,x_2,...,x_{n_x} \}$, then, using set notaiton,<br>
$A = \{(μ_A(x_i)/x_i|x_i \in X, i = 1,...,n_x\}$

Using sum notation,<br>
$A = μ_A(x_1)/x_1 + μ_A(x2)/x_2 + ... | μ_A(x_{n_x})/x_{n_x} = ∑_{i_1}^{n_x} μ_A(x_i)/x_i$

where the sum should not be confused with algebraic summation. The use of sum notation above sumply serves as an indication that A is a set of odered pairs. A continuous fuzzy set, $A$, is denoted as:<br>
$A= ∫_X μ(x)/x$


**III. Membership Function**

Membership functions for fuzzy sets can be of any shape of type as determined by experts in the domain over which the sets are defined. While desingers of fuzzy sets have much freedom in selecting appropriate membership functions, these functions must satisfy the following constraints:
- A membership function must be bounded from below by 0 and from ablove by 1.
- The range of a membership function must therefore be [0,1].
- For each $x \in X,μ_A(x)$ must be unique. That is, the same element cannot map to different degrees of membership for the same fuzzy set.

**1. Triangular function**

$trianglmf(x;a,b,c) = max(min( \frac{x-a}{b-c}, \frac{c-x}{c-b},0)$

**2. Trapezo function**

$trapezomf(x;a,b,c,d) = max(min( \frac{x-a}{b-a},1, \frac{d-x}{d-c},0)$

**3. Gaussian function**

$gaussmf(x;σ,c) = e^{-\frac{1}{2}(\frac{x-c}{σ})^2}$

**4. Logistic function**

$logisticmf(x;a,b) = \frac{1}{1+e^{-b(x-a)}}$


```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Triangular membership function
def triangular(x, a, b, c):
    return max(0, min((x - a) / (b - a), (c - x) / (c - b)))

# Trapezoidal membership function
def trapezoidal(x, a, b, c, d):
    return max(0, min((x - a) / (b - a), 1, (d - x) / (d - c)))

# Gaussian membership function
def gaussian(x, mean, sigma):
    return np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))

# Logistic membership function
def logistic(x, a, b):
    return 1 / (1 + np.exp(-b * (x - a)))

# Generate x values
x_values = np.linspace(0, 10, 1000)

# Triangular function parameters
a_tri = 2
b_tri = 5
c_tri = 8

# Trapezoidal function parameters
a_trap = 2
b_trap = 4
c_trap = 7
d_trap = 9

# Gaussian function parameters
mean_gaussian = 5
sigma_gaussian = 1.5

# Logistic function parameters
a_logistic = 5
b_logistic = 0.5

# Calculate y values for each membership function
y_triangular = [triangular(x, a_tri, b_tri, c_tri) for x in x_values]
y_trapezoidal = [trapezoidal(x, a_trap, b_trap, c_trap, d_trap) for x in x_values]
y_gaussian = [gaussian(x, mean_gaussian, sigma_gaussian) for x in x_values]
y_logistic = [logistic(x, a_logistic, b_logistic) for x in x_values]

# Plot the membership functions
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x_values, y_triangular, label='Triangular')
plt.title('Triangular Membership Function')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, y_trapezoidal, label='Trapezoidal')
plt.title('Trapezoidal Membership Function')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x_values, y_gaussian, label='Gaussian')
plt.title('Gaussian Membership Function')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x_values, y_logistic, label='Logistic')
plt.title('Logistic Membership Function')
plt.legend()

plt.tight_layout()
plt.show()
```

**IV. Fuzzy Operators**

As for crisp sets, relations and operations are defined for fuzzy sets. Each of these realations and operators are defined below. For this purpose let $X$ be the domain, or universe, and $A$ and $B$ are fuzzy sets defined over the domain $X$.

**Equality**: Two fuzzy setss $A$ and $B$ are equal if and only if the sets have the same domains, and $μ_A(x) = μ_B(x)$ for all $x \in X$. That is, $A = B$

**Containment of fuzzy sets**: For two-valued sets, $A ⊂ B$ if all the elements of $A$ are also elements of $B$. For fuzzy sets, this definition is not complete, and the degree of membership of elements to the sets have to be considered.

Fuzzy set $A$ is a subset of fuzzy set $B$ if and only if $μ_A(x) ≤ μ_B(x)$ for all $x \in X$. That is, $A ⊂ B$.

**Complement of a fuzzy set (NOT)**: The complement of a two-valued set is sumply the set containing the entire domain without the elements of that set. For fuzzy sets, the complement of the set $A$ consists of all the elements of set $A$, but the membership degree differ. Let $̅A$ denote the complement of set A. Then, for all $x \in X, μ_{̅A}(x) = 1 - μ_A(x)$. It í also follows that $A ∩ ̅A \neq Ø$ and $A ∪ ̅A \neq X$.

**Intersection of fuzzy sets (AND)**: The intersection of two-valued sets is the set of elements occuring in both sets. Operators that implement intersection are refered as **t-norms**. The result of a t-norm is a set that contain all the elements of the two fuzzy sets, but with degree of membership that depends on the specific t-norm. A number of t-norm have been used, of which the min-operator and the product operator are the most popular. If $A$ and $B$ are two fuzzy sets, then:<br>
- **Min-operator**: $μ_{A∩B}(x) = min \{ μ_A(x),μ_B(x)\}, ∀x \in X$.
- **Product operator**: $μ_{A∩B}(x) = μ_A(x)μ_B(x), ∀x \in X$.

The difference between the two operations should be noted. Taking the product
of membership degrees is a much stronger operator than taking the minimum,
resulting in lower membership degrees for the intersection. It should also be
noted that the ultimate result of a series of intersections approaches 0.0, even if the degrees of memberships to the original sets are high.

**Union of fuzzy sets (OR)**: The union of two-valued sets contains the elements of all of the sets. The same is true for fuzzy sets, but with membership degrees that depend on the specific union operator used. These operators are referred to as **s-norms**, of which the max-operator and summation operator are most frequently used:<br>
- **Max-operator**: $μ_{A∪B}(x) = max \{ μ_A(x),μ_B(x)\}, ∀x \in X$.
- **Summation operator**: $μ_{A∪B}(x) = μ_A(x) + μ_B(x) - μ_A(x)μ_B(x), ∀x \in X$

Again, careful consideration must be given to the differences between the two approaches above. In the limit, a series of unions will have a result that approximates 1.0, even though membership degrees are low for the original sets.


**V. Fuzzy Set Characteristics**

**Normality**: A fuzzy set $A$ is normal if that set has an element that belongs to the set with degree 1. That is,<br>
$∃x \in A • μ_A(x) = 1$

then $A$ is normal, otherwise, $A$ is subnormal:<br>
$sup_xμ_A(x) = 1$

**Height**: The height of a fuzzy set is defined as the supremum of the membership function.<br>
$height(A) = sup_xμ_A(x)$

**Support**: The support of fuzzy set $A$ is the set of all elements in the universe of discource, $X$, that belongs to $A$ with non-zero membership. That is,<br>
$support(A) = \{ x \in X | μ_A(x) > 0 \}$

**Core**: The core of fuzzy set $A$ is the set of all elements in the domain that belongs to $A$ with membership degree 1. That is,<br>
$core(A)  = \{ x \in X | μ_A(x) = 1 \}$

**Alpha-cut**: The set of elements of $A$ with membership degree greater than $α$ is referred to as the $α-cut$ of $A$.<br>
$A_α = \{ x \in X| μ_A(x) ≥ α \}$

**Unimodality**: A fuzzy set is unimodal if its membership function is a unimodal function, i.e the function has just one maximum.

**Cardinality**: The cardinality of two-valued sets is suimply the number of elements within the sets. This is not the same for fuzzy sets. The cardinality of fuzzy set $A$, for a finite domain, $X$, is defined as:<br>
$card(A) = ∑_{x \in X} μ_A(x)$

and for an infinite domain,<br>
$card(A) = ∫_{x \in X}μ_A(x)dx$

For example: if $X = \{a,b,c,d \}$ and $A = 0.3/a + 0.9/b + 0.1/c + 0.7/d$, then $card(A) = 0.3 + 0.9 + 0.1 +0.7 = 2.0$.

**Normalization**: A fuzzy set is normalized by dividing the membership function by the height of the fuzzy set. That is,<br>
$normalized(A) = \frac{μ_A(x)}{height(x)}$

- $card(A) + card(B) = card(A∪B) + card(A∩B)$

- $card(A) + card(̅A) = card(X)$
